```python
# Inicialização grafo (n: número de vértices, m: número de arcos )
n, m = (int(tmp) for tmp in input().split(" "))
adj = [[] for _ in range(n)]

for k in range(m):
  i, j = (int(tmp) for tmp in input().split(" "))
  adj[i].append(j)

# Inicialização dos vértices
BRANCO = 1
CINZA = 2
PRETO = 3

lista_tempos_iniciais = [0] * n
lista_tempos_finais = [0] * n
lista_cores_vertices = [0] * n

# Método visita DFS
def visitaDfs(v):
    global tempo 
    tempo += 1
    lista_tempos_iniciais[v] = tempo
    lista_cores_vertices[v] = CINZA

    print(f"({v}", end=" ")

    for u in adj[v]:
      if (lista_cores_vertices[u] == BRANCO):
        visitaDfs(u)
    tempo += 1
    lista_tempos_finais[v] = tempo
    lista_cores_vertices[v] = PRETO
    print(f"{v})", end=" ") 

# Inicio do DFS
for v in range(n):
  lista_cores_vertices[v] = BRANCO
tempo = 0

for v in range(n):
  if(lista_cores_vertices[v] == BRANCO):
    visitaDfs(v)
     
```
